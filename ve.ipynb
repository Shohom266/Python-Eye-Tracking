```python
import time

try:
    from neurapy.robot import Robot
except Exception:
    Robot = None


def dump_state(r, label):
    print(f"\n--- {label} ---", flush=True)

    for name in ("program_status", "motion_status", "get_errors"):
        if hasattr(r, name):
            try:
                value = getattr(r, name)()
                print(f"{name}: {value}", flush=True)
            except Exception as e:
                print(f"{name}: ERROR {e}", flush=True)

    try:
        print(f"automatic: {r.is_robot_in_automatic_mode()}", flush=True)
    except Exception as e:
        print(f"automatic: ERROR {e}", flush=True)

    try:
        print(f"joint angles: {r.get_current_joint_angles()}", flush=True)
    except Exception as e:
        print(f"joint angles: ERROR {e}", flush=True)


def test_joint_jog(joint=1, direction=1, hold_cycles=300, cycle_sleep=0.01):
    if Robot is None:
        raise RuntimeError("NeuraPy is not available")

    r = Robot()
    try:
        dump_state(r, "before")

        if not r.is_robot_in_automatic_mode():
            raise RuntimeError("Robot is not in Automatic mode")

        print("init_program()", flush=True)
        r.init_program()
        dump_state(r, "after init_program")

        jog_velocity = [0.0] * 6
        jog_velocity[joint - 1] = 1 if direction >= 0 else -1.0
        print(f"turn_on_jog(jog_velocity={jog_velocity}, jog_type='Joint')", flush=True)
        r.turn_on_jog(jog_velocity=jog_velocity, jog_type="Joint")
        dump_state(r, "after turn_on_jog")

        for i in range(max(1, int(hold_cycles))):
            print(f"jog loop {i+1}/{hold_cycles}", flush=True)
            r.jog(set_jogging_external_flag=1)
            time.sleep(cycle_sleep)

        dump_state(r, "after jog loop")

    except Exception as e:
        print(f"TEST ERROR: {e}", flush=True)
        dump_state(r, "after error")
        raise
    finally:
        try:
            print("turn_off_jog()", flush=True)
            r.turn_off_jog()
        except Exception as e:
            print(f"turn_off_jog ERROR: {e}", flush=True)

        try:
            print("stop()", flush=True)
            r.stop()
        except Exception as e:
            print(f"stop ERROR: {e}", flush=True)

        dump_state(r, "final")


if __name__ == "__main__":
    test_joint_jog(joint=1, direction=1)
```
